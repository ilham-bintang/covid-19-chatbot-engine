```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import tensorflow_text

import pandas as pd
```

```python
module = hub.load('https://tfhub.dev/google/universal-sentence-encoder-qa/3')
```

```python
df = pd.read_csv('data.csv', delimiter=',')
df
```

```python
response = df["Response"].values
context = df["Context"].values
```

```python
len(response), len(context)
```

```python
response_embeddings = module.signatures['response_encoder'](
        input=tf.constant(response),
        context=tf.constant(context))
```

```python
def predict(question):
    question_embeddings = module.signatures['question_encoder'](tf.constant([question]))
    res = np.inner(question_embeddings['outputs'], response_embeddings['outputs'])
    answer = response[np.argmax(res[0])]
    confidence = res[0][np.argmax(res[0])]
    return answer, confidence
```

```python
ans, conf = predict("How coronavirus spreaded?")
```

```python
ans, conf
```

```python
while True:
    question = input(">>> ")
    ans, conf = predict(question)
    if conf <= 0.05:
        print("🤖 : I am not sure what's the answer. Can you spell it clearly?")
    else:
        print("🤖 : "+ ans)
        print("Confidence: " + str(conf))
    print("\n")
```
